```python
katok = []
select = 0

def add_data(data):
    katok.append(None)
    a = len(katok)
    katok[a-1] = data
    if(a != 1):
        for i in range(a):
            m_index = i
            for j in range(i,a):
                if katok[j][1] > katok[m_index][1]:
                    m_index = j
            katok[i] = katok[m_index]
            katok[m_index] = katok[i]
            
def delete_data(data):
    
    del katok[data]
    
while(True):
    select = int(input("선택하세요(1.추가, 2.삭제, 3.종료)"))
    if(select == 1):
        data = input("추가할 데이터--> ")
        add_data(data)
        print(katok)
    elif(select == 2):
        data = int(input("삭제할 순서--> "))
        delete_data(data)
        print(katok)
    elif(select == 3):
        exit
```
